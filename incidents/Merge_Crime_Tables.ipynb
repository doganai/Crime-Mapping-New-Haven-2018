```python
import pandas as pd
```

```python
incident_list = pd.read_csv(r'incidents.csv',index_col=2)
incident_types = pd.read_csv(r'incident_types.csv',index_col=2)

incident_list
```

```python
mergeDf = pd.merge(incident_types, incident_list, left_on = ['IncidentTypes','IncidentTypes'], right_on = ['Incident Type','Incident Type'])
```

```python
mergeDf
```

```python
#Convert to CSV
mergeDf.to_csv('incidentMerged.csv')
```
